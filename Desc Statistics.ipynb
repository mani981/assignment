# data represented in bars

```python
import matplotlib.pyplot as plt
country=['usa','india','goa','uk','france']
GDP_Per_Capita=[48000,38000,500000,45000,440000]
plt.bar(country,GDP_Per_Capita)
plt.title('Country vs GDP_Per_Capita')
plt.xlabel('country')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
plt.barh(country,GDP_Per_Capita)
plt.show()
```

```python
plt.bar(country,GDP_Per_Capita,color='r')
plt.show()
```
