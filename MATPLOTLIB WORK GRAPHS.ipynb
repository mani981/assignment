```python
import pandas as pd
import matplotlib.pyplot as plt  

df=pd.read_csv("data.csv")
monthList=df['month no'].tolist()
faceCremSalesData=df ['facecream'].tolist()
faceWashSalesData=df ['facewash'].tolist()
toothPastSalesData=df ['toothpast'].tolist()
bathingsoapSalesData=df ['bathingsoap'].tolist()
shampooSalesData=df ['shampoo'].tolist()
moisturizerSalesData=df ['moisturizer'].tolist()

plt.plot(monthList,faceCremSalesData,label='Face cream Sales Data',marker='o',linewidth=3)
plt.plot(monthList,faceWashSalesData,label='Face Wash Sales Data',marker='o',linewidth=3)
plt.plot(monthList,toothPastSalesData,label='Tooth Past Sales Data',marker='o',linewidth=3)
plt.plot(monthList,bathingsoapSalesData,label='Tooth Past Sales Data',marker='o',linewidth=3)
plt.plot(monthList,shampooSalesData,label='Tooth Past Sales Data',marker='o', linewidth=3)
plt.plot(monthList,moisturizerSalesData,label='Tooth Past Sales Data',marker='o', linewidth=3)

plt.xlabel('month no')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(monthList)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("data.csv")
monthList=df['month no'].tolist()
facecreamsalesdata=df['facecream'].tolist()
facewashsalesdata=df['facewash'].tolist()
plt.bar([a-0.25 for a in monthList],facecreamsalesdata,width=0.25,label='face cream sales data',align='edge')
plt.bar([a+0.25 for a in monthList],facewashsalesdata,width=-0.25,label='face wash sales data',align='edge')
plt.xlabel('month no')
plt.ylabel('sales units in number')
plt.legend(loc='upper left')
plt.xticks(monthList)
plt.grid(True,linewidth=1,linestyle="--")
plt.title('Facewash and facecream sales data')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("data.csv")
monthList=df['month no'].tolist()
facecreamsalesdata=df['facecream'].tolist()
facewashsalesdata=df['facewash'].tolist()
toothpastsalesdata=df['toothpast'].tolist()
bathingsoapsalesdata=df['bathingsoap'].tolist()
shampoosalesdata=df['shampoo'].tolist()
moisturizersalesdata=df['moisturizer'].tolist()
plt.plot([],[],color='m',label='facecream',linewidth=5)
plt.plot([],[],color='c',label='facewash',linewidth=5)
plt.plot([],[],color='r',label='toothpast',linewidth=5)
plt.plot([],[],color='k',label='bathingsoap',linewidth=5)
plt.plot([],[],color='g',label='shampoo',linewidth=5)
plt.plot([],[],color='y',label='moisturizer',linewidth=5)
plt.stackplot(monthList,facecreamsalesdata,facewashsalesdata,toothpastsalesdata,bathingsoapsalesdata,shampoosalesdata,moisturizersalesdata,colors=['m','c','r','k','g','y'])
plt.xlabel('month no')
plt.ylabel('sales units in number')
plt.title('all product sales data using stack plot')
plt.show()
```
